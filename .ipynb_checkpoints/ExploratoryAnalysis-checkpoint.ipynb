All sessions were summarized in data/SessionsInfo.pickle.

Individual session, trial infos were in data/*.pickle, named with corresponding subjects' unique id.


```python
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from numpy import concatenate as concat
%matplotlib inline

with open("data/SessionsInfo.pickle", "rb") as f:
    SessionsInfo = pickle.load(f)
```

### Overview

```python
SessionsInfo
```

```python
fig, axs = plt.subplots(1, 4, figsize = (25,5))
sns.histplot(ax = axs[0], data = SessionsInfo, x = 'SessionDuration_min', kde = True, alpha=.5)
sns.histplot(ax = axs[1],data = SessionsInfo, x = 'Ntrials', kde =  False, alpha=.5)
sns.histplot(ax = axs[2], data = SessionsInfo, x = 'rewardRate', kde = True, alpha=.5)
SessionsInfo['Country'].value_counts().plot(ax = axs[3], kind='bar', alpha=.5, ylabel = 'Count')
```

### individual

```python
with open("data/data_A17P1A5OQHAECM_10061827.pickle","rb") as f:
    df = pickle.load(f)
df
```

```python
fig, axs = plt.subplots(2,1,figsize=(25, 10))
axs[0].plot(df.mainTrials, df.IDport, 'k.-' )
axs[0].plot(df.mainTrials, df.IDchoice,'g.' )
error_ind = df.Reward==0
axs[0].plot(df.loc[error_ind,'mainTrials'], df.loc[error_ind,'Switch'], 'r.')
axs[1].plot(df['mainTrials'], 1e3*df['RT'], 'k-')
axs[1].plot(df.loc[error_ind,'mainTrials'], 1e3+df.loc[error_ind,'Switch'], 'r.')
axs[1].set_xlabel('Trial')
axs[1].set_ylabel('RT(msec)')
axs[1].set_yscale('log')

fig, axs = plt.subplots(1, 2, figsize=(25, 5)) 
sns.histplot(ax = axs[0], data = df, x = 'RT', hue = 'Reward', kde = True, common_norm= False, alpha=.5, linewidth=0)
axs[0].set_xlim(0, 2)
sns.kdeplot(ax = axs[1], data = df, x = 'RT',y = 'fixDuration', hue = 'Reward', fill = False, alpha = 0.5)
axs[1].set_xlim(0, 2)

plt.figure(figsize=(25,8))
trialIndex = df.index
for Index in trialIndex:
    fix, choice = df.loc[Index, 'mousetracking_fix'], df.loc[Index, 'mousetracking_choice']
    breakpoint = choice[-1,:]
    x, y, t = 0, 1, 2
    plt.scatter(x = concat((fix[:,x], choice[:,x]),axis = 0),
                y = concat((fix[:,y], choice[:,y]),axis = 0),
                c = concat((fix[:,t], choice[:,t]),axis = 0), s = 10, alpha = 0.3)
    plt.scatter(x = breakpoint[0], y = breakpoint[1],s=50, edgecolors='r', facecolors = 'none')
plt.axis('equal')
plt.show()
```

```python
from shapely.geometry import Point
import geopandas as gpd
from geopandas import GeoDataFrame

index = SessionsInfo['Latitude'].map(lambda x: True if type(x)==float else False)
dfgeo = SessionsInfo.loc[index, ['Longitude','Latitude']]

geometry = [Point(xy) for xy in zip(dfgeo['Longitude'], dfgeo['Latitude'])]
gdf = GeoDataFrame(dfgeo, geometry=geometry)   

#this is a simple map that goes with geopandas
world = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))
gdf.plot(ax=world.plot(figsize=(10, 6)), marker='o', color='red', markersize=15);
```
